```python
import numpy as np
import pandas as pd
from scipy.stats import binom
import matplotlib.pyplot as plt

# Parámetros de la distribución binomial
n_trials = 10
p_success = 0.6
n_experimentos = 1000

# Simulación de los datos: número de éxitos en 1000 experimentos binomiales
datos_simulados = np.random.binomial(n=n_trials, p=p_success, size=n_experimentos)
df = pd.DataFrame(datos_simulados, columns=["Éxitos"])

# *********** COMPLEMENTO DEL EJERCICIO 1 ************** #

# Histograma de las ventas diarias
plt.hist(datos_simulados, bins=range(min(datos_simulados), max(datos_simulados) + 2),
         color='skyblue', edgecolor='black', align='left')
plt.title("Distribución Binomial: Intentos exitos por experimento")
plt.xlabel("Numero de exitos")
plt.ylabel("Frecuencia")
plt.xticks(range(min(datos_simulados), max(datos_simulados) + 1))
plt.show()

# 2. Cálculo de la varianza teórica
varianza_teorica = n_trials * p_success * (1 - p_success)

# 2. Cálculo de la varianza empírica
varianza_empirica = df["Éxitos"].var()

# 4. Cálculo de los momentos (empíricos)
momentos = {}
for k in range(1, 5):
    momentos[k] = np.mean(df["Éxitos"]**k)

# 5. Función generadora de momento (MGF)
def mgf(t, n, p):
    return (1 - p + p * np.exp(t))**n

# Calcular la MGF para varios valores de t
t_values = np.linspace(-1, 1, 100)
mgf_values = mgf(t_values, n_trials, p_success)

# 6. Esperanza matemática (media teórica)
esperanza_teorica = n_trials * p_success
esperanza_empirica = df["Éxitos"].mean()

# 7. Función característica
def func_caracteristica(t, n, p):
    return (1 - p + p * np.exp(1j * t))**n

# Calcular la función característica para varios valores de t
func_carac_values = func_caracteristica(t_values, n_trials, p_success)

# Mostrar los resultados
print(f"Varianza Teórica: {varianza_teorica}")
print(f"Varianza Empírica: {varianza_empirica}")
print(f"Momentos Empíricos: {momentos}")
print(f"Esperanza Teórica: {esperanza_teorica}")
print(f"Esperanza Empírica: {esperanza_empirica}")

# Graficar la MGF y la Función Característica
plt.figure(figsize=(12, 6))

# MGF
plt.subplot(1, 2, 1)
plt.plot(t_values, mgf_values, label="MGF Teórica", color="blue")
plt.title("Función Generadora de Momentos (MGF)")
plt.xlabel("t")
plt.ylabel("MGF")
plt.grid(True)

# Función Característica
plt.subplot(1, 2, 2)
plt.plot(t_values, np.real(func_carac_values), label="Función Característica (Parte Real)", color="red")
plt.title("Función Característica")
plt.xlabel("t")
plt.ylabel("Parte Real de la Función Característica")
plt.grid(True)

plt.tight_layout()
plt.show()
```
